```python
# ## Region 1: South Cascade Glacier
# ## Upper left (lon/lat): -121.0742643433662, 48.37159557493268
# ## Lower right (lon/lat): -121.03683181392472, 48.341195823749906

# lon_up = -121.0742643433662
# lat_up =  48.37159557493268
# lon_lo = -121.03683181392472
# lat_lo =  48.341195823749906
# region_name='South Cascade Glacier'

## Region 2: Maryland
## Upper left (lon/lat): -77.41454197206069, 39.290044188370636
## Lower right (lon/lat): -77.39562385238138, 39.239373031026304


lon_up = -77.41454197206069
lat_up =  39.290044188370636
lon_lo = -77.39562385238138
lat_lo =  39.239373031026304
region_name='Maryland'

#Set the shape data dir path
shape_filename='/home/jovyan/CmCt/data/ne_10m_coastline/ne_10m_coastline.shp'

```

# Configure ICESat-2 API and Region of Interest

```python
# Configure ICESat-2 API
icesat2.init("slideruleearth.io")
# Configure Region of Interest
region = [ {"lon":lon_up, "lat": lat_up},
           {"lon":lon_lo, "lat": lat_up},
           {"lon":lon_lo, "lat": lat_lo},
           {"lon":lon_up, "lat": lat_lo},
           {"lon":lon_up, "lat": lat_up} ]
```

# ATL06

```python
# Build ATL06 Request
 #calculates geolocated elevations in the selected region at a 20m along-track resolution.
parms_alt = {
    "poly": region,
    "srt": icesat2.SRT_LAND,
    "cnf": icesat2.CNF_SURFACE_HIGH,
    "ats": 10.0,
    "cnt": 10,
    "len": 40.0,
    "res": 20.0
}
# Request ATL06 Data
gdf6 = icesat2.atl06p(parms_alt)
# gedi04a[gedi04a["3dep.value"].notna()]
```

## ATL08

```python
# # Request ATL08 Data
# parms_alt08 = {
#     "poly": region,
#     "srt": 0,#srt=0 (land)
#     "cnf": 0,#cnf=0 (no native filtering)
#     "ats": 10.0,
#     "cnt": 0,
#     "len": 40.0,
#     "res": 20.0
# }

# Request ATL08 Data
parms_alt08 ={"asset": "icesat2",
        "phoreal": {},
        "srt": -1,#-1,#0
        "poly":region}

gdf8= icesat2.atl08p(parms_alt08)
```

# GEDI

```python
## GEDI
gedi.init("slideruleearth.io", verbose=True)
geojson = earthdata.tnm(short_name='Digital Elevation Model (DEM) 1 meter', polygon=region)
parms = {
    "poly": region,
    "degrade_flag": 0,
    "l2_quality_flag": 1,
    "samples": {"3dep": {"asset": "usgs3dep-1meter-dem", "catalog": geojson}} 
}

# Request GEDI Data
gedi04a = gedi.gedi04ap(parms) 
# Massage DataFrame: trim NaN and no-data rows and flatten samples
gdf_3dep = gedi04a[gedi04a["3dep.value"].notna()]
def getFirstValue(x):
    if type(x["3dep.value"]) == float:
        return x['3dep.value']
    else:
        return x['3dep.value'][0]
gdf_3dep = gdf_3dep.copy()        
gdf_3dep["3dep"] = gdf_3dep.apply(lambda x: getFirstValue(x), axis=1)
gdf_3dep = gdf_3dep[gdf_3dep["3dep"] > -9999.0]
```

```python
def all_plot_variables(var_label, region_name, region, gdf6, gdf8, gdf_3dep,gdf6_var_name,gdf8_var_name,gdf_3dep_var_name):
    # Calculate Extent for the "zoomed" plots
    lons = [p["lon"] for p in region]
    lats = [p["lat"] for p in region]
    lon_margin = (max(lons) - min(lons)) * 0.1
    lat_margin = (max(lats) - min(lats)) * 0.1
    
    # Prepare region outline
    box_lon = [e["lon"] for e in region]
    box_lat = [e["lat"] for e in region]

    # Create a 2x2 grid for subplots
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
    ax1, ax2, ax3, ax4 = axes[0][0], axes[0][1], axes[1][0], axes[1][1]
    
    ############################################################################
    # 1) Zoomed Track for gdf6 (top-left)
    ############################################################################
    ax1.set_title(f"Zoomed Tracks of {var_label} (ATL06)")
    
    # Plot using GeoPandas
    scatter = gdf6.plot(ax=ax1, column=gdf6_var_name, cmap="coolwarm", s=1.0, zorder=3)
    
    # Set axis limits and aspect
    ax1.set_xlim(min(lons) - lon_margin, max(lons) + lon_margin)
    ax1.set_ylim(min(lats) - lat_margin, max(lats) + lat_margin)
    ax1.set_aspect('equal', adjustable='box')

    # Colorbar for gdf6
    norm_6 = Normalize(vmin=gdf6[gdf6_var_name].min(), vmax=gdf6[gdf6_var_name].max())
    sm_6 = ScalarMappable(norm=norm_6, cmap="coolwarm")
    sm_6.set_array([])
    cbar_6 = fig.colorbar(sm_6, ax=ax1, fraction=0.046, pad=0.04, shrink=0.8)
    cbar_6.set_label(var_label)

    # Draw the region outline
    ax1.plot(box_lon, box_lat, linewidth=1.5, color='r', zorder=2)
    
    ############################################################################
    # 2) Zoomed Track for gdf8 (top-right)
    ############################################################################
    ax2.set_title(f"Zoomed Tracks of {var_label} (ATL08)")
    
    scatter = gdf8.plot(ax=ax2, column=gdf8_var_name, cmap="coolwarm", s=1.0, zorder=3)
    
    ax2.set_xlim(min(lons) - lon_margin, max(lons) + lon_margin)
    ax2.set_ylim(min(lats) - lat_margin, max(lats) + lat_margin)
    ax2.set_aspect('equal', adjustable='box')

    # Colorbar for gdf8
    norm_8 = Normalize(vmin=gdf8[gdf8_var_name].min(), vmax=gdf8[gdf8_var_name].max())
    sm_8 = ScalarMappable(norm=norm_8, cmap="coolwarm")
    sm_8.set_array([])
    cbar_8 = fig.colorbar(sm_8, ax=ax2, fraction=0.046, pad=0.04, shrink=0.8)
    cbar_8.set_label(var_label)

    # Region outline
    ax2.plot(box_lon, box_lat, linewidth=1.5, color='r', zorder=2)
    
    ############################################################################
    # 3) Zoomed Track for gdf_3dep (bottom-left)
    ############################################################################
    ax3.set_title(f"Zoomed Tracks of {var_label} (3DEP)")
    
    scatter = gdf_3dep.plot(ax=ax3, column=gdf_3dep_var_name, cmap="coolwarm", s=1.0, zorder=3)
    
    ax3.set_xlim(min(lons) - lon_margin, max(lons) + lon_margin)
    ax3.set_ylim(min(lats) - lat_margin, max(lats) + lat_margin)
    ax3.set_aspect('equal', adjustable='box')

    # Colorbar for gdf_3dep (zoomed)
    norm_3dep = Normalize(vmin=gdf_3dep[gdf_3dep_var_name].min(), vmax=gdf_3dep[gdf_3dep_var_name].max())
    sm_3dep = ScalarMappable(norm=norm_3dep, cmap="coolwarm")
    sm_3dep.set_array([])
    cbar_3dep = fig.colorbar(sm_3dep, ax=ax3, fraction=0.046, pad=0.04, shrink=0.8)
    cbar_3dep.set_label(var_label)

    # Region outline
    ax3.plot(box_lon, box_lat, linewidth=1.5, color='r', zorder=2)

    ############################################################################
    # 4) Basemap plot of gdf_3dep only (bottom-right)
    ############################################################################
    ax4.set_title(f"Basemap of {var_label} (3DEP) over {region_name}")
    
    # Reproject to EPSG:4326
    gdf_3dep_4326 = gdf_3dep.to_crs(epsg=4326)
    
    # Create Basemap
    margin_map = 2.0#3.0  # degrees margin
    m = Basemap(
        projection="lcc",
        resolution="h",#"h"
        lat_0=39.29,
        lon_0=-77.40,
        llcrnrlon=gdf_3dep_4326.geometry.x.min() - margin_map,
        llcrnrlat=gdf_3dep_4326.geometry.y.min() - margin_map,
        urcrnrlon=gdf_3dep_4326.geometry.x.max() + margin_map,
        urcrnrlat=gdf_3dep_4326.geometry.y.max() + margin_map,
        ax=ax4,
    )
    
    # Draw map features
    m.shadedrelief()
    m.drawcoastlines(color="gray")
    m.drawcountries(color="gray")
    m.drawstates(color="gray")

    # Transform coordinates
    xs, ys = m(gdf_3dep_4326.geometry.x.values, gdf_3dep_4326.geometry.y.values)

    # Scatter plot on Basemap
    sc = ax4.scatter(
        xs, ys,
        c=gdf_3dep_4326[gdf_3dep_var_name],
        cmap="coolwarm",
        s=20.0,
        zorder=3
    )
    
    # Colorbar for basemap
    # (Reuse the same normalization as used for the 3dep data)
    sm_map = ScalarMappable(norm=norm_3dep, cmap="coolwarm")
    sm_map.set_array([])
    cbar_map = fig.colorbar(sm_map, ax=ax4, fraction=0.046, pad=0.04)
    cbar_map.set_label(var_label)

    # Adjust layout and show
    plt.tight_layout()
    plt.show()
```

```python
var_label='elevation'
gdf6_var_name="h_mean"
gdf8_var_name="h_te_median"
gdf_3dep_var_name="3dep"


all_plot_variables(var_label, region_name, region, gdf6, gdf8, gdf_3dep,gdf6_var_name,gdf8_var_name,gdf_3dep_var_name)
```

```python
def plot_variable_by_year(gdf8,variable, region,data_prod):
    """
    Plots h_te_median from gdf8 as points within the bounding region,
    coloring each year's data differently.
    """

    #Bounding box for the "zoomed" region
    lons = [p["lon"] for p in region]
    lats = [p["lat"] for p in region]
    lon_margin = (max(lons) - min(lons)) * 0.1
    lat_margin = (max(lats) - min(lats)) * 0.1
    

    #figure and axes
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.set_title(f"Zoomed Tracks of {data_prod} {variable} by Year ")
    

    # Separate data by year
    # Extract unique years from the datetime index
    unique_years = sorted(gdf8.index.year.unique())
    
    # Define a colormap or a list of colors
    # Here tab10 colormap is used for 10 distinct colors.
    # If thereare more than 10 years, use a bigger colormap or manually define a list.
    cmap = colormaps["tab10"].resampled(len(unique_years))
    scale_factor=0.5
    # Plot each year's data with a unique color
    for i, year in enumerate(unique_years):
        # Filter rows for this particular year
        subset = gdf8[gdf8.index.year == year]
        
        # Plot using GeoPandas
        #  - We do NOT specify `column=` here, because were coloring by year, not by a numeric range.
      
        subset.plot(
            ax=ax,
            marker='o',
            color=cmap(i),  # color depends on the year
            label=str(year),  # legend label
            markersize=5,   
            zorder=3
        )

    
    box_lon = [pt["lon"] for pt in region]
    box_lat = [pt["lat"] for pt in region]
    ax.plot(box_lon, box_lat, linewidth=1.5, color='red', zorder=2)

    ax.set_xlim(min(lons) - lon_margin, max(lons) + lon_margin)
    ax.set_ylim(min(lats) - lat_margin, max(lats) + lat_margin)
    ax.set_aspect('equal', adjustable='box')

    ax.legend(title="Year", loc="best")

    plt.tight_layout()
    plt.show()
```

```python
gdf6_var_name="h_mean"
data_prod6="ATL06"
plot_variable_by_year(gdf6,gdf6_var_name, region,data_prod6)


gdf8_var_name="h_te_median"
data_prod8="ATL08"
plot_variable_by_year(gdf8,gdf8_var_name, region,data_prod8)

gdf_3dep_var_name="3dep"
data_prod_dep="GEDI"
plot_variable_by_year(gdf_3dep,gdf_3dep_var_name, region,data_prod_dep)
```
