# GEDI

```python
## GEDI
gedi.init("slideruleearth.io", verbose=True)
geojson = earthdata.tnm(short_name='Digital Elevation Model (DEM) 1 meter', polygon=region)
parms = {
    "poly": region,
    "degrade_flag": 0,
    "l2_quality_flag": 1,
    "samples": {"3dep": {"asset": "usgs3dep-1meter-dem", "catalog": geojson}} 
}

# Request GEDI Data
gedi04a = gedi.gedi04ap(parms) 
# Massage DataFrame: trim NaN and no-data rows and flatten samples
gdf_3dep = gedi04a[gedi04a["3dep.value"].notna()]
def getFirstValue(x):
    if type(x["3dep.value"]) == float:
        return x['3dep.value']
    else:
        return x['3dep.value'][0]
gdf_3dep = gdf_3dep.copy()        
gdf_3dep["3dep"] = gdf_3dep.apply(lambda x: getFirstValue(x), axis=1)
gdf_3dep = gdf_3dep[gdf_3dep["3dep"] > -9999.0]
```

## Difference 

```python
# Avoid SettingWithCopyWarning by using .loc properly
import geopandas as gpd
from scipy.spatial import cKDTree


def calculate_altO_3dep_difference(gdf_filtered,gdf_3dep_filtered,variable):

    # Ensure we are working with a copy of the dataframes to avoid modifying slices
    gdf_filtered = gdf_filtered.copy()
    gdf_3dep_filtered = gdf_3dep_filtered.copy()
    
    # Use .loc to prevent SettingWithCopyWarning
    gdf_filtered.loc[:, 'x'] = gdf_filtered.geometry.x
    gdf_filtered.loc[:, 'y'] = gdf_filtered.geometry.y
    gdf_3dep_filtered.loc[:, 'x'] = gdf_3dep_filtered.geometry.x
    gdf_3dep_filtered.loc[:, 'y'] = gdf_3dep_filtered.geometry.y
    
    # Build KDTree for nearest-neighbor search
    tree = cKDTree(gdf_3dep_filtered[['x', 'y']].to_numpy())
    distances, indices = tree.query(gdf_filtered[['x', 'y']].to_numpy(), k=1)
    
    # Assign nearest neighbor values without modifying slices directly
    gdf_filtered = gdf_filtered.assign(nearest_3dep=gdf_3dep_filtered.iloc[indices]['3dep'].values)
    
    # Compute topography difference
    gdf_filtered = gdf_filtered.assign(
        elev_diff=gdf_filtered[variable] - gdf_filtered['nearest_3dep']
    )
    
    # Select relevant columns for display
    result_df = gdf_filtered[['geometry', variable, 'nearest_3dep', 'elev_diff']]
    
    return result_df 

```

```python
#Filter data to get geometry and elevation variable
gdf6_filtered = gdf6[['geometry', 'h_mean']]
gdf8_filtered = gdf8_htemedian_filtered[['geometry', 'h_te_median']]
gdf_3dep_filtered = gdf_3dep[['geometry', '3dep']]


# Calculate difference
gdf6_var_name="h_mean"
result_df_alt06=calculate_altO_3dep_difference(gdf6_filtered,gdf_3dep_filtered,gdf6_var_name)


gdf8_var_name="h_te_median"
result_df_alt08=calculate_altO_3dep_difference(gdf8_filtered,gdf_3dep_filtered,gdf8_var_name)


gdf_3dep['elev_diff'] = gdf_3dep['elevation'] - gdf_3dep['3dep']
```

## Plots

```python
def all_plot_variables(var_label, region_name, region, gdf6, gdf8, gdf_3dep,gdf6_var_name,gdf8_var_name,gdf_3dep_var_name,result_df_alt06,result_df_alt08):
   
    # Calculate Extent for the "zoomed" plots
    lons = [p["lon"] for p in region]
    lats = [p["lat"] for p in region]
    lon_margin = (max(lons) - min(lons)) * 0.1
    lat_margin = (max(lats) - min(lats)) * 0.1
    
    # Prepare region outline
    box_lon = [e["lon"] for e in region]
    box_lat = [e["lat"] for e in region]

    # Create a 2x3 grid for subplots
    fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))
    
    # Explicitly assign each subplot to variables
    ax1, ax2, ax3 = axes[0][0], axes[0][1], axes[0][2]
    ax4, ax5, ax6 = axes[1][0], axes[1][1], axes[1][2]


    
    # Function to plot each dataset
    def plot_dataset(ax, gdf, col_name, title, norm, sm,cmap_var, outline=True):
        ax.set_title(title)
        scatter = gdf.plot(ax=ax, column=col_name, cmap=cmap_var, norm=norm, s=1.0, zorder=3)
        ax.set_xlim(min(lons) - lon_margin, max(lons) + lon_margin)
        ax.set_ylim(min(lats) - lat_margin, max(lats) + lat_margin)
        ax.set_aspect('equal', adjustable='box')
        cbar = fig.colorbar(sm, ax=ax, fraction=0.046, pad=0.04, shrink=0.8)
        cbar.set_label(var_label)
        if outline:
            ax.plot(box_lon, box_lat, linewidth=1.5, color='r', zorder=2)

    

    ## Colorbar for product

    # Compute global min and max across all datasets for color mapping
    global_vmin = min(
        gdf6[gdf6_var_name].min().min(),
        gdf8[gdf8_var_name].min().min(),
        gdf_3dep[var_label].min().min()
    )

    print(global_vmin)
    
    global_vmax = max(
        gdf6[gdf6_var_name].max().max(),
        gdf8[gdf8_var_name].max().max(),
        gdf_3dep[var_label].max().max()
    )

    print(global_vmax)
    norm_var = Normalize(vmin=global_vmin, vmax=global_vmax)
    sm_var = ScalarMappable(norm=norm_var, cmap="viridis")
    sm_var.set_array([])

    # Select cmap
    cmap_prod="viridis"


    # Plot each dataset
    plot_dataset(ax1, gdf6, gdf6_var_name, f"Zoomed Tracks of{region_name}\n {gdf6_var_name} (ATL06)", norm_var, sm_var,cmap_prod)
    plot_dataset(ax2, gdf8, gdf8_var_name, f"Zoomed Tracks of {region_name}\n {gdf8_var_name} (ATL08)", norm_var, sm_var,cmap_prod)
    plot_dataset(ax3, gdf_3dep, gdf_3dep_var_name, f"Zoomed Tracks of{region_name}\n {var_label} (3DEP)", norm_var, sm_var,cmap_prod)



   # Colorbar for Difference Calculation
    # Compute the absolute max across all datasets for color mapping
    global_vmax_diff = max(
        abs(result_df_alt06['elev_diff'].min()),
        abs(result_df_alt08['elev_diff'].min()),
        abs(gdf_3dep['elev_diff'].min()),
        abs(result_df_alt06['elev_diff'].max()),
        abs(result_df_alt08['elev_diff'].max()),
        abs(gdf_3dep['elev_diff'].max())
    )
    
    # Set the symmetric min and max values
    global_vmin_diff = -global_vmax_diff

    
    # norm_diff = Normalize(vmin=global_vmin_diff, vmax=global_vmax_diff)
    # norm_diff = Normalize(vmin=-100, vmax=100)
    norm_diff = Normalize(vmin=-25, vmax=25)
    sm_diff = ScalarMappable(norm=norm_diff, cmap="coolwarm")
    sm_diff.set_array([])
    # Select cmap
    cmap_diff="coolwarm"

    # Plot differences
    plot_dataset(ax4, result_df_alt06, 'elev_diff', f"Zoomed Tracks of {region_name}\n ATL06 ({gdf6_var_name})-3DEP", norm_diff, sm_diff,cmap_diff)
    plot_dataset(ax5, result_df_alt08, 'elev_diff', f"Zoomed Tracks of {region_name}\nATL08 ({gdf8_var_name})-3DEP", norm_diff, sm_diff,cmap_diff)
    plot_dataset(ax6, gdf_3dep, 'elev_diff', f"Zoomed Tracks of{region_name}\n GEDI ({var_label})-3DEP", norm_diff, sm_diff,cmap_diff)

    # Adjust layout and show
    plt.tight_layout()
    plt.show()

  
```

```python
def plot_variable_by_year(gdf8,variable, region,data_prod,region_name):
    """
    Plots h_te_median from gdf8 as points within the bounding region,
    coloring each year's data differently.
    """

    #Bounding box for the "zoomed" region
    lons = [p["lon"] for p in region]
    lats = [p["lat"] for p in region]
    lon_margin = (max(lons) - min(lons)) * 0.1
    lat_margin = (max(lats) - min(lats)) * 0.1
    

    #figure and axes
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.set_title(f"Zoomed Tracks of {region_name} {data_prod} {variable} by Year ")
    

    # Separate data by year
    # Extract unique years from the datetime index
    unique_years = sorted(gdf8.index.year.unique())
    
    # Define a colormap or a list of colors
    # Here tab10 colormap is used for 10 distinct colors.
    # If thereare more than 10 years, use a bigger colormap or manually define a list.
    cmap = colormaps["tab10"].resampled(len(unique_years))
    scale_factor=0.5
    # Plot each year's data with a unique color
    for i, year in enumerate(unique_years):
        # Filter rows for this particular year
        subset = gdf8[gdf8.index.year == year]
        
        # Plot using GeoPandas
        #  - We do NOT specify `column=` here, because we’re coloring by year, not by a numeric range.
      
        subset.plot(
            ax=ax,
            marker='o',
            color=cmap(i),  # color depends on the year
            label=str(year),  # legend label
            markersize=5,   
            zorder=3
        )

    
    box_lon = [pt["lon"] for pt in region]
    box_lat = [pt["lat"] for pt in region]
    ax.plot(box_lon, box_lat, linewidth=1.5, color='red', zorder=2)

    ax.set_xlim(min(lons) - lon_margin, max(lons) + lon_margin)
    ax.set_ylim(min(lats) - lat_margin, max(lats) + lat_margin)
    ax.set_aspect('equal', adjustable='box')

    ax.legend(title="Year", loc="best")

    plt.tight_layout()
    plt.show()
```

```python
gdf6_var_name="h_mean"
data_prod6="ATL06"
plot_variable_by_year(gdf6,gdf6_var_name, region,data_prod6,region_name)


gdf8_var_name="h_te_median"
data_prod8="ATL08"
plot_variable_by_year(gdf8_htemedian_filtered,gdf8_var_name, region,data_prod8,region_name)

gdf_3dep_var_name="3dep"
data_prod_dep="GEDI"
plot_variable_by_year(gdf_3dep,gdf_3dep_var_name, region,data_prod_dep,region_name)
```

```python
from ipyleaflet import Map, basemaps, Rectangle, CircleMarker, ImageOverlay
from ipywidgets import Layout
import matplotlib.colors as colors
import matplotlib.pyplot as plt
import io
import base64
from PIL import Image

def all_plot_variables_basemap(var_label, region_name, region, gdf6, gdf8, gdf_3dep, 
                               gdf6_var_name, gdf8_var_name, gdf_3dep_var_name, 
                               result_df_alt06, result_df_alt08):

    def generate_colorbar(norm_data, cmap_data):
        """Creates a colorbar image and returns it as a base64-encoded URL (no file saving)."""
        
        fig = plt.figure(figsize=(1.5, 6))  # Adjust figure size
        ax = fig.add_axes([0.05, 0.05, 0.2, 0.9])  # Define an axis for colorbar
    
        # Create colorbar using ScalarMappable
        sm = plt.cm.ScalarMappable(norm=norm_data, cmap=cmap_data)
        cb = plt.colorbar(sm, cax=ax)
    
        # Define tick locations at actual min, mid, and max values
        tick_values = [norm_data.vmin, (norm_data.vmin + norm_data.vmax) / 2, norm_data.vmax]
        
        # Apply actual tick values and labels
        cb.ax.set_yticks(tick_values)  
        cb.ax.set_yticklabels([f"{tick_values[0]:.2f}", f"{tick_values[1]:.2f}", f"{tick_values[2]:.2f}"])
        
        # Set label for colorbar with proper spacing
        cb.set_label('h_mean (meters)', fontsize=14, labelpad=15)
        
        # Adjust layout to prevent text cutoff
        plt.tight_layout()
    
        # Save colorbar image to memory (no file saving), with a white background
        buf = io.BytesIO()
        plt.savefig(buf, format='png', bbox_inches='tight', dpi=300, facecolor='white')  
        plt.close(fig)
    
        # Convert the image buffer to a base64-encoded string
        buf.seek(0)
        encoded_image = base64.b64encode(buf.read()).decode("utf-8")
    
        # Return the image as a data URL
        return f"data:image/png;base64,{encoded_image}"



    def plot_leaflet_scatter(gdf, var_name, norm_data, cmap_data):
        """Create an interactive ipyleaflet scatter plot with global color normalization and a colorbar overlay."""

        # Compute bounding box
        lon_up, lat_up = gdf.geometry.x.max(), gdf.geometry.y.max()
        lon_lo, lat_lo = gdf.geometry.x.min(), gdf.geometry.y.min()
        center_lat, center_lon = (lat_up + lat_lo) / 2, (lon_up + lon_lo) / 2

        # Create ipyleaflet map
        m = Map(
            basemap=basemaps.Esri.WorldImagery,
            center=(center_lat, center_lon),
            zoom=14,
            layout=Layout(width="600px", height="600px")
        )

        # Add a rectangle to highlight the region
        rectangle = Rectangle(
            bounds=((lat_lo, lon_lo), (lat_up, lon_up)),
            color="red",
            weight=2,
            fill_opacity=0.2
        )
        m.add_layer(rectangle)

        # Add scatter markers with globally normalized color scale
        for _, row in gdf.iterrows():
            lat, lon = row.geometry.y, row.geometry.x
            h_value = row[var_name]

            rgba = cmap_data(norm_data(h_value))
            hex_color = colors.rgb2hex(rgba[:3])

            marker = CircleMarker(
                location=(lat, lon),
                radius=3,  
                color=hex_color,
                fill=True,
                fill_color=hex_color,
                fill_opacity=0.8
            )
            m.add_layer(marker)

        # Generate the colorbar image as a base64 URL
        colorbar_url = generate_colorbar(norm_data, cmap_data)

        
        # Compute bounding box dimensions
        map_height = lat_up - lat_lo
        map_width = lon_up - lon_lo
        
        # Define a proportional width for the colorbar
        # colorbar_width = map_width * 0.15  # Adjust this multiplier for better fitting
        colorbar_width = map_width * 0.3  # Adjust this multiplier for better fitting
        
        # Adjust bounds to maintain proper aspect ratio
        colorbar_bounds = ((lat_lo, lon_up + 0.02), (lat_up, lon_up + colorbar_width))   

        # Add colorbar overlay to the map
        colorbar_overlay = ImageOverlay(
            url=colorbar_url,
            # bounds=((lat_lo, lon_up), (lat_up, lon_up + 0.05)) 
            bounds=colorbar_bounds  # Moves further right# Position colorbar
        )
        m.add_layer(colorbar_overlay)

        return m

    # Compute global min and max across all datasets for color mapping
    global_vmin = min(
        gdf6[gdf6_var_name].min(),
        gdf8[gdf8_var_name].min(),
        gdf_3dep[gdf_3dep_var_name].min()
    )
    
    global_vmax = max(
        gdf6[gdf6_var_name].max(),
        gdf8[gdf8_var_name].max(),
        gdf_3dep[gdf_3dep_var_name].max()
    )

    # Normalize color scale
    # norm = colors.Normalize(vmin=global_vmin, vmax=global_vmax)
    # cmap_var = plt.colormaps.get_cmap("viridis")  


    # norm = colors.Normalize(vmin=1591.1436, vmax=3303.5697757524563)#South Cascade Glacier
    norm = colors.Normalize(vmin=49.887447, vmax=289.18048)#Maryland

    cmap_var = plt.colormaps.get_cmap("viridis")     

    # Show ipyleaflet scatter plots with consistent color mapping
    print("\n**Interactive ipyleaflet Maps with Global Color Normalization & Colorbar:**")
    display(plot_leaflet_scatter(gdf6, gdf6_var_name, norm, cmap_var))
    display(plot_leaflet_scatter(gdf8, gdf8_var_name, norm, cmap_var))
    display(plot_leaflet_scatter(gdf_3dep, gdf_3dep_var_name, norm, cmap_var))

    # Normalize for difference plots
    norm_diff = colors.Normalize(vmin=-25, vmax=25)
    cmap_diff = plt.colormaps.get_cmap("coolwarm") 
    
    display(plot_leaflet_scatter(result_df_alt06, 'elev_diff', norm_diff, cmap_diff))
    display(plot_leaflet_scatter(result_df_alt08, 'elev_diff', norm_diff, cmap_diff))
    display(plot_leaflet_scatter(gdf_3dep, 'elev_diff', norm_diff, cmap_diff))

# Run the function
all_plot_variables_basemap(var_label, region_name, region, gdf6, gdf8, gdf_3dep, 
                           gdf6_var_name, gdf8_var_name, gdf_3dep_var_name, 
                           result_df_alt06, result_df_alt08)
```
