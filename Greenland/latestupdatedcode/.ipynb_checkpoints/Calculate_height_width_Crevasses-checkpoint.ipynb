```python
#UTM region

utmXmin =575900  # Easting (meters east of central meridian)
utmYmin = 7671600 # Northing (meters north of equator)
utmXmax = 576600
utmYmax = 7672800


# Convert UTM to lat/lon using pyproj with UTM Zone 22N for Jakobshavn Isbr√¶ region
transformer_jakobshavn = Transformer.from_crs("EPSG:32622",  "EPSG:4326", always_xy=True)

# Transform bounding box corners
lon_min, lat_min = transformer_jakobshavn.transform(utmXmin, utmYmin)
lon_max, lat_max = transformer_jakobshavn.transform(utmXmax, utmYmax)

# # Output the range
# lat_min, lat_max, lon_min, lon_max

lon_up = lon_max
lat_up = lat_max
lon_lo = lon_min
lat_lo = lat_min

# Configure Region of Interest
region = [ {"lon":lon_up, "lat": lat_up},
           {"lon":lon_lo, "lat": lat_up},
           {"lon":lon_lo, "lat": lat_lo},
           {"lon":lon_up, "lat": lat_lo},
           {"lon":lon_up, "lat": lat_up} ]
```

```python
# Configure ICESat-2 API
icesat2.init("slideruleearth.io")

length=30.0
res=10.0
# Build ATL06 Request
parms = {
    "poly": region,
    "srt": 3,       # land ice
    "cnf":-1 ,#4,-1
    "ats": 20.0,    # 20m along-track spacing
    "cnt": 10,
    "len": length,
    "res": res
}
gdf6 = icesat2.atl06p(parms)
```

```python
cycle=3
subset = gdf6[
    (gdf6['gt'] == 50) &
    (gdf6['rgt'] == 666) &
    (gdf6['spot'] == 5) &
    (gdf6['cycle'] == cycle)
]

strong_beam = gdf6[
    (gdf6['gt'] == 50) &
    (gdf6['rgt'] == 666) &
    (gdf6['spot'] == 5) &
    (gdf6['cycle'] == cycle)
]

weak_beam = gdf6[
    (gdf6['gt'] == 60) &
    (gdf6['rgt'] == 666) &
    (gdf6['spot'] == 6) &
    (gdf6['cycle'] == cycle)
]
```

```python
# Plot h_mean vs x_atc
plt.figure(figsize=(10, 6))
plt.plot(subset['x_atc'], subset['h_mean'], marker='o', linestyle='-', markersize=3)
plt.xlabel('x_atc')
plt.ylabel('h_mean')
plt.title(f'h_mean vs x_atc (cycle {cycle})')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Sort by along-track coordinate
subset = subset.sort_values('x_atc').reset_index(drop=True)
# subset
```

```python
# Step 1: Find all the local minima points

# Initialize mask of False
subset['is_bottom'] = False

# Convert to numpy array for efficient indexing
h_vals = subset['h_mean'].values

# Identify local minima (excluding endpoints)
local_minima_indices = np.where((h_vals[1:-1] < h_vals[:-2]) & (h_vals[1:-1] < h_vals[2:]))[0] + 1

# Mark these indices in the DataFrame
subset.loc[subset.index[local_minima_indices], 'is_bottom'] = True

# Get corresponding DataFrame indices
bottom_points = list(subset.index[local_minima_indices])

# Mark these indices in the DataFrame
subset.loc[bottom_points, 'is_bottom'] = True

```

```python
# Plot
plt.figure(figsize=(12, 6))
plt.plot(subset['x_atc'], subset['h_mean'], label='h_mean', alpha=0.5)
plt.scatter(subset.loc[subset['is_bottom'], 'x_atc'],
            subset.loc[subset['is_bottom'], 'h_mean'],
            color='red', label='Crevasse bottoms')
plt.xlabel('x_atc')
plt.ylabel('h_mean')
plt.title('Detected Local Minima as Crevasse Bottoms')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Step 3: Extract crevasse depth using surface fit

# Detect local maxima (surface points)
h_vals = subset['h_mean'].values
x_vals = subset['x_atc'].values

surface_indices = np.where((h_vals[1:-1] > h_vals[:-2]) & (h_vals[1:-1] > h_vals[2:]))[0] + 1
surface_points = [(x_vals[i], h_vals[i]) for i in surface_indices]

# Fit surface spline if enough points
if len(surface_points) >= 4:
    surface_points = sorted(surface_points, key=lambda x: x[0])
    surface_df = pd.DataFrame(surface_points, columns=['x', 'h'])
    surface_df = surface_df.drop_duplicates(subset='x').sort_values('x')

    x_surface = surface_df['x'].values
    h_surface = surface_df['h'].values

    if len(x_surface) >= 4:
        spline = CubicSpline(x_surface, h_surface)
        subset['surface_fit'] = spline(subset['x_atc'])
        subset['depth'] = subset['surface_fit'] - subset['h_mean']
    else:
        subset['surface_fit'] = np.nan
        subset['depth'] = np.nan
else:
    subset['surface_fit'] = np.nan
    subset['depth'] = np.nan

```

```python
# Step 4: Apply depth threshold
depth_threshold = 2.0  # meters
subset['is_valid_crevasse'] = (subset['is_bottom']) & (subset['depth'] >= depth_threshold)
```

```python
# Plot detected crevasse bottom points and surface fit
plt.figure(figsize=(12, 6))
plt.plot(subset['x_atc'], subset['h_mean'], label='h_mean', alpha=0.5)
if subset['surface_fit'].notna().any():
    plt.plot(subset['x_atc'], subset['surface_fit'], label='Surface fit', linestyle='--')
plt.scatter(subset.loc[subset['is_valid_crevasse'], 'x_atc'],
            subset.loc[subset['is_valid_crevasse'], 'h_mean'],
            color='red', label='Crevasse bottoms')
plt.xlabel('x_atc')
plt.ylabel('Elevation (m)')
plt.title('Crevasse Depth and Surface Fit')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Step 5: Estimate crevasse width 
# First, find all local maxima (surface points)
surface_x_vals = subset.iloc[surface_indices]['x_atc'].values

# Estimate width for each valid crevasse
widths = []

for i in subset[subset['is_valid_crevasse']].index:
    center_x = subset.loc[i, 'x_atc']
    
    # Find nearest surface peaks on either side
    left_edges = surface_x_vals[surface_x_vals < center_x]
    right_edges = surface_x_vals[surface_x_vals > center_x]
    
    edge_left = left_edges.max() if left_edges.size > 0 else np.nan
    edge_right = right_edges.min() if right_edges.size > 0 else np.nan
    
    width = edge_right - edge_left if pd.notnull(edge_right) and pd.notnull(edge_left) else np.nan
    widths.append(width)

# Assign widths to the valid crevasse points
subset.loc[subset['is_valid_crevasse'], 'width'] = widths

```

```python
# Show final table of detected crevasses
crevasses_df = subset[subset['is_valid_crevasse']][['x_atc', 'h_mean', 'depth', 'width']]
print("Detected crevasses:")
print(crevasses_df)
```

 Summary of workflow:

Find minima ‚Üí crevasse bottom
Detect surface points as local maxima (i.e., crevasse edges).
Fit a CubicSpline to those local maxima.
Interpolate the surface height at all x_atc.
Subtract h_mean from the surface fit to get depth.
Filter by depth threshold

Approximate width using left/right edges
Plot results and show detected crevasses



Handles duplicate x_surface values

Checks for at least 4 unique points before fitting the spline (minimum for cubic spline)

Avoids empty window errors

Produces clean plot and table summary
